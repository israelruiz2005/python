# Encontre as raizes, o vertice e o grafio da equação do 2 grau

$$y=x^2 - 6x + 5$$

```python
from sympy import *
from sympy.plotting import *
```

```python
a = int(input('Entre com valor para a: '))
```

```python

b = int(input('Entre com valor para b:'))
```

```python
y = Function('y')
x = Symbol('x')
```

$$ y = x^2 -a*x + b$$

```python

y = x**2 -a*x + b
```

```python
solve(y,x)
```
